## Conductivity vs. Carrier concentration

### Case 1: ZnO:X

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import itertools
from IPython.display import HTML
%matplotlib inline
```

```python
#specify path to data
path_to_data = os.getcwd()+"/data/ZnO/"
#get file names in directory
csvFiles = os.listdir(path_to_data)
#initialize empty list to catch any none CSV files
pdfFile = []
#create a figure and axis
fig = plt.figure(figsize=(8, 6), dpi=80)
axis = fig.add_subplot(111)
#create a cycle of markers for data points

marker = itertools.cycle((u'o', u'>', u'v', u's', u'D')) 
#loop through files and plot data
for File in csvFiles:
    try:
        #load data from file
        points = np.loadtxt(path_to_data + File, delimiter=",")
        if File =='Linear fit.csv':
            axis.plot(points[:,0],1/points[:,1], marker="None", markersize=4, color='black')
        else:
            axis.plot(points[:,0],1/points[:,1], label=File[:-4], marker=marker.next(), markersize=6, linestyle="None")
    except:
        pdfFile.append(File)

#set plot properteis
axis.set_xlim([1e14,1e21])
axis.set_ylim([1e-4,1e4])
axis.set_ylabel("Conductivity, $(1/\\Omega-cm)$")
axis.set_xlabel("Carrier concentration $(cm^{-3})$")
axis.set_yscale("log")
axis.set_xscale("log")
plt.rcParams.update({'font.size': 14})
plt.legend(prop={'size':12},loc="upper left");
```
