## Conductivity vs. Carrier concentration

```python
#setup and import statements
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import itertools
import scipy.optimize as so
from scipy import stats
from IPython.display import HTML, Latex
%matplotlib inline

#define fitting function
def fitfunc(x, b, a):
    return a*(x - nc) + b
```

### Case 1: ZnO:X thin films on sapphire at room temperature

```python
#specify path to data
path_to_data = os.getcwd()+"/data/ZnO/"
#get file names in directory
csvFiles = os.listdir(path_to_data)
#initialize empty list to catch any none CSV files
pdfFile = []
#create a figure and axis
fig = plt.figure(figsize=(8, 6), dpi=80)
axis = fig.add_subplot(111)
#create a cycle of markers for data points

marker = itertools.cycle((u'o', u'>', u'v', u's', u'D')) 
#loop through files and plot data
for File in csvFiles:
    try:
        #load data from file
        points = np.loadtxt(path_to_data + File, delimiter=",")
        if File =='Linear fit.csv':
            axis.plot(points[:,0],1/points[:,1], marker="None", markersize=4, color='black')
        else:
            axis.plot(points[:,0],1/points[:,1], label=File[:-4], marker=marker.next(), markersize=6, linestyle="None")
    except:
        pdfFile.append(File)

#set plot properties
axis.set_xlim([1e14,1e21])
axis.set_ylim([1e-4,1e4])
axis.set_ylabel("Conductivity, $(1/\\Omega-cm)$")
axis.set_xlabel("Carrier concentration $(cm^{-3})$")
axis.set_yscale("log")
axis.set_xscale("log")
plt.rcParams.update({'font.size': 14})
plt.legend(prop={'size':12},loc="upper left");
```

<b>Source:</b> Ellmer, Klaus, Andreas Klein, and Bernd Rech. Transparent conductive zinc oxide: basics and applications in thin film solar cells. Vol. 104. Springer, 2007. (page 323)

### Case 2: boron doped diamond at 10K

```python
#system specific constants
nc = 4.5
xScale = 10**20

#specify path to data
path_to_data = os.getcwd()+"/data/boron doped diamond/"
#get file names in directory
csvFiles = os.listdir(path_to_data)
#initialize empty list to catch any none CSV files
pdfFile = []
#create a figure and axis
fig = plt.figure(figsize=(8, 6), dpi=80)
axis = fig.add_subplot(111)

#loop through files and plot data
for File in csvFiles:
    try:
        #load data from file
        points = np.loadtxt(path_to_data + File, delimiter=",")
        if File == "Experimental data.csv":
            #add points to plot
            axis.plot(points[:,0]*xScale,1/points[:,1], label=File[:-4] + ", T=10K", marker="o", markersize=4)
            #slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(points[:,0]), np.log(1/points[:,1]))
            pars, covar = so.curve_fit(fitfunc, np.log(points[:,0]), np.log(1/points[:,1]))
            x = np.log(np.linspace(points[:,0].min(),points[:,0].max(), 1000))
            #y = (1/slope)*x + intercept
            y = fitfunc(x, *pars)
            print "Form of power-law fit:\n\ty = %G*(x - %G)^%G" % (pars[0]*xScale,nc*xScale,pars[1])
            axis.plot(np.exp(x)*xScale,np.exp(y), label="Power-law fit")
        else:
            #add points to plot
            axis.plot(points[:,0]*xScale,1/points[:,1], label=File[:-4], marker="o", markersize=4)
    except:
        pdfFile.append(File)
#set plot properties
axis.set_xlim([.2e20,1.1e22])
axis.set_ylim([.01,1.4e4])
axis.set_ylabel("Conductivity, $(1/\\Omega-cm)$")
axis.set_xlabel("Carrier concentration $(cm^{-3})$")
axis.set_yscale("log")
axis.set_xscale("log")
plt.rcParams.update({'font.size': 14})
plt.legend(prop={'size':12},loc="upper left");
```

<b>Source:</b> Klein, Thierry, et al. "Metal-insulator transition and superconductivity in boron-doped diamond." Physical Review B 75.16 (2007): 165313.
