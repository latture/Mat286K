## Conductivity vs. Carrier concentration

### Case 1: CuGaTe<sub>2</sub>

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#specify path to data
path_to_data = os.getcwd()+"/data/CuGaTe2/"
#get file names in directory
csvFiles = os.listdir(path_to_data)
#initialize empty list to catch any none CSV files
pdfFile = []
#create a figure and axis
fig = plt.figure()
axis = fig.add_subplot(111)

#loop through files and plot data
for File in csvFiles:
    try:
        #load data from file
        points = np.loadtxt(path_to_data + File, delimiter=",")
        #add points to plot
        axis.plot(points[:,0],points[:,1], label=File[:-4])
        axis.plot(points[:,0],points[:,1], marker="o", markersize=4)
    except:
        pdfFile.append(File)

#set plot properteis
axis.set_xlim([1e18,1e20])
axis.set_ylabel("Conductivity, $(1/\\Omega m)$")
axis.set_xlabel("p (holes,cm$^3$)")
axis.set_yscale("log")
axis.set_xscale("log")
plt.rcParams.update({'font.size': 14})
plt.legend(prop={'size':12},loc="upper left")
```
